**1. Write a program that asks the user for the number of days. The program then prints
out the number of seconds in the number of days given.**

```python
%%writefile no1.cpp
#include <iostream>
using namespace std;

int main(){
  int days;

  cout << "Enter the number of days: ";
  cin >> days;

  //converting days to sec
  int sec = days * 86400;

  cout << "Number of seconds in " << days << " day(s) is: " << sec << endl;

  return 0;
}
```

```python
!g++ no1.cpp -o no1
!./no1
```

**2. Write a program that asks a user to input the radius then the program calculates the
volume of a sphere (the formula for the volume is (4/3) πr3). Use the inbuilt
exponential function in C++ to compute (r3).**

```python
%%writefile no2.cpp
#include <iostream>
#include <cmath>
using namespace std;

int main(){
  double radius;

  cout << "Enter the radius of the sphere: ";
  cin >> radius;

  //calculating volume
  const double pi = 3.14159265;
  double volume = (4.0 / 3.0 ) * pi * pow(radius, 3);

  cout << "The volume of sphere radius = " << radius << " is " << volume << endl;

  return 0;
}
```

```python
!g++ no2.cpp -o no2
!./no2
```

**3. Using functions, write a program to compute the area and perimeter of a square.
The program should ask the user to enter a number corresponding to the side length
of the square and display the area and perimeter of the square.**

```python
%%writefile no3.cpp
#include <iostream>
using namespace std;

//area fnc
double calculateArea(double side){
  return side * side;
}

//perimeter fnc
double calculatePerimeter(double side){
  return 4 * side;
}

int main(){
  double side;

  cout << "Enter side length: ";
  cin >> side;

  double area = calculateArea(side);
  double perimeter = calculatePerimeter(side);

  cout << "The area of the square with side length " << side << " is " << area << endl;
  cout << "The perimeter of the square with side legth " << side << " is " << perimeter << endl;

  return 0;
}
```

```python
!g++ no3.cpp -o no3
!./no3
```

**4. Write a program using functions that determines whether a character input by a
user is uppercase or lower case.**

```python
%%writefile no4.cpp
#include <iostream>
#include <cctype>
using namespace std;

//fnc to check if uppercase or lowercase
void checkCase(char ch) {
    if (isupper(ch)) {
        cout << ch << " is an Uppercase letter." << endl;
    }
    else if (islower(ch)) {
        cout << ch << " is a Lowercase letter." << endl;
    }
    else {
        cout << ch << " invalid." << endl;
    }
}

int main(){
  char ch;

  cout << "Enter the letter: ";
  cin >> ch;

  checkCase(ch);

  return 0;
}
```

```python
!g++ no4.cpp -o no4
!./no4
```

**5. The following is pseudocode for a program being designed. Write the C++ program
equivalent of the same.**<br>
BEGIN <br>
SET x TO 0, y TO 20 <br>
REPEAT <br>
SUBTRACT 4 FROM y <br>
ADD 2/y TO x <br>
UNTIL <br>
y IS LESS THAN 6 <br>
DISPLAY x <br>
END

```python
%%writefile no5.cpp
#include <iostream>
using namespace std;

int main(){
  double x = 0;
  double y = 20;

  do{
    y = y - 4;
    x = y + (2.0 / y);
  } while (
    y >= 6
  );

  cout << "Final value of x: " << x << endl;

  return 0;
}
```

```python
!g++ no5.cpp -o no5
!./no5
```

**6. Write a C++ program that does the following:<br>
i.Uses a loop for a user to continually input 5 values to populate an array. <br>
ii.Calculates and displays the average of the values input into the array.**

```python
%%writefile no6.cpp
#include <iostream>
using namespace std;

int main(){
  int k = 0;
  double sum = 0;
  int value[5];
  do{
    cout << "Enter value: ";
    cin >> value[k];

    sum = sum + value[k];
    k = k + 1;
  } while ( k < 5);

  int average = sum / 5;

  cout << "The average of the values is: " << average << endl;

  return 0;
}
```

```python
!g++ no6.cpp -o no6
!./no6
```
